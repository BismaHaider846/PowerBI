```python
# prompt: remove the outliers from data

def remove_outliers(df, column_name):
  """
  This function removes outliers from a specific column in a DataFrame.

  Args:
      df: The DataFrame containing the data.
      column_name: The name of the column to remove outliers from.

  Returns:
      A new DataFrame with outliers removed from the specified column.
  """

  # Calculate the interquartile range (IQR)
  Q1 = df[column_name].quantile(0.25)
  Q3 = df[column_name].quantile(0.75)
  IQR = Q3 - Q1

  # Define the lower and upper bounds for outlier detection
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR

  # Filter the DataFrame to exclude outliers
  df_filtered = df[((df[column_name] >= lower_bound) & (df[column_name] <= upper_bound))]

  return df_filtered

# Remove outliers from the Sales column
df_filtered = remove_outliers(df, 'Sales')

# Remove outliers from the Quantity column
df_filtered = remove_outliers(df_filtered, 'Quantity')

# Remove outliers from the Profit column
df_filtered = remove_outliers(df_filtered, 'Profit')

# Print the new DataFrame with outliers removed
print(df_filtered)
```

```python
# prompt: apply the simple moving average

# Calculate the simple moving average (SMA) of the Sales column with a window size of 3
df_filtered['SMA_Sales'] = df_filtered['Sales'].rolling(window=3).mean()

# Print the DataFrame with the added SMA column
print(df_filtered)
```

```python
# prompt: apply the linear regression model

from sklearn.linear_model import LinearRegression

# Create a linear regression model
model = LinearRegression()

# Define the dependent and independent variables
X = df_filtered[['Sales', 'Quantity']]
y = df_filtered['Profit']

# Train the model
model.fit(X, y)

# Print the model coefficients and intercept
print(f"Model coefficients: {model.coef_}")
print(f"Model intercept: {model.intercept_}")

# Make predictions using the trained model
y_pred = model.predict(X)

# Print the predicted values
print(f"Predicted values: {y_pred}")
```

```python
# prompt: apply multiple linear regression by taking y as sales independent variable and x as the profit dependent variable print and load

import pandas as pd
import statsmodels.api as sm

# Load the data
df = pd.read_excel("/content/7-Eleven - Superstore Sales Data.xlsx")

# Define the dependent and independent variables
y = df['Sales']
X = df['Profit']

# Add a constant term to the independent variable
X = sm.add_constant(X)

# Create the linear regression model
model = sm.OLS(y, X)

# Fit the model
results = model.fit()

# Print the model summary
print(results.summary())

# Make predictions using the trained model
y_pred = results.predict(X)

# Print the predicted values
print(f"Predicted values: {y_pred}")
```

```python
# prompt: interpret the above multiple regression model print and load

# Print the model summary
print(results.summary())

# Interpret the model coefficients
print("Coefficient of Profit:", results.params[1])

# Interpret the intercept
print("Intercept:", results.params[0])

# Interpret the R-squared value
print("R-squared:", results.rsquared)

# Interpret the adjusted R-squared value
print("Adjusted R-squared:", results.rsquared_adj)

# Interpret the F-statistic
print("F-statistic:", results.fvalue)

# Interpret the p-value
print("p-value:", results.pvalues[1])
```

R-squared and adjusted R- square: the r-square value is 0.230 indicates that 21.3% of variability in the dependent variable can be explained by the model . the adjusted r-square  which penelizes the addition of irrelevant features is also 0.230 indicating a good balance between complexity and predictive power.
F-STATISTIC AND PROB (F-STATISTUC): the f-statistic is 2976.24 and corresponding probability is 0.00 this suggests that the model as a whole is statistically significant and atleast one of the predictors is relevant in predicting the dependent variable.

```python
# prompt: apply the exponential moving average on the data

# Calculate the exponential moving average (EMA) of the Sales column with a window size of 3 and alpha value of 0.5
df_filtered['EMA_Sales'] = df_filtered['Sales'].ewm(alpha=0.5, adjust=False).mean()

# Print the DataFrame with the added EMA column
print(df_filtered)
```

```python
# prompt: create scatter plot of features vs median house value

import matplotlib.pyplot as plt

# Create a scatter plot of the Sales feature against the median house value
plt.scatter(df_filtered['Sales'], df_filtered['Profit'])
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Scatter Plot of Sales vs Profit')
plt.show()

# Create a scatter plot of the Quantity feature against the median house value
plt.scatter(df_filtered['Quantity'], df_filtered['Profit'])
plt.xlabel('Quantity')
plt.ylabel('Profit')
plt.title('Scatter Plot of Quantity vs Profit')
plt.show()
```

```python
# prompt: calculate the mean squared error of simple moving average and exponential moving average

# Calculate the mean squared error (MSE) of the simple moving average (SMA)
sma_errors = df_filtered['Sales'] - df_filtered['SMA_Sales']
sma_mse = sma_errors.pow(2).mean()

# Calculate the mean squared error (MSE) of the exponential moving average (EMA)
ema_errors = df_filtered['Sales'] - df_filtered['EMA_Sales']
ema_mse = ema_errors.pow(2).mean()

# Print the MSE values
print(f"MSE of SMA: {sma_mse}")
print(f"MSE of EMA: {ema_mse}")
```

```python
# prompt: apply the arima model to the data

import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Define the model order
p = 2
d = 1
q = 1

# Create the model
model = ARIMA(df_filtered['Sales'], order=(p, d, q))

# Fit the model
results = model.fit()

# Print the model summary
print(results.summary())

# Make predictions using the trained model
forecast_horizon = 12
y_pred = results.forecast(steps=forecast_horizon)

# Print the predicted values
print(f"Predicted values: {y_pred}")

# Plot the actual and predicted values
plt.plot(df_filtered['Sales'], label='Actual')
plt.plot(y_pred, label='Predicted')
plt.xlabel('Time')
plt.ylabel('Sales')
plt.title('ARIMA Model Forecast')
plt.legend()
plt.show()
```

```python
# prompt: calculate the mean square error of arima model

# Calculate the mean squared error (MSE) of the ARIMA model
arima_errors = df_filtered['Sales'] - y_pred
arima_mse = arima_errors.pow(2).mean()

# Print the MSE value
print(f"MSE of ARIMA: {arima_mse}")
```

```python
# prompt: compare the performance of SMA EMA AND ARIMA model based on their MSE

# Compare the MSE values of the three models
print("MSE of SMA:", sma_mse)
print("MSE of EMA:", ema_mse)
print("MSE of ARIMA:", arima_mse)

# Based on the MSE values, the model with the lowest MSE performs best
if sma_mse < ema_mse and sma_mse < arima_mse:
  print("The SMA model performs best.")
elif ema_mse < sma_mse and ema_mse < arima_mse:
  print("The EMA model performs best.")
else:
  print("The ARIMA model performs best.")
```
